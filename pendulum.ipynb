```python
import pendulum
```

```python
# The library natively supports the RFC 3339 format,
# most ISO 8601 formats and some other common formats.

xs = [
    '2020-12-18 22:16:32.842 +0200',
    '2020-12-18 20:16:32.842 +0000',
    '2020-12-18 20:16:32.842',
    '2020-12-18T20:16:32.842',
    '2020-12-18T20:16:32.842+0000',
    '2020-12-18T20:16:32.842+00:00',
    '2020-12-18T20:16:32.842+0200',
    '2020-12-18T20:16:32.842+02:00',
    '2020-12-18T20:16:32.842 +0000',
    '2020-12-18T20:16:32.842 +00:00',
    '2020-12-18T20:16:32.842 +0200',
    '2020-12-18T20:16:32.842 +02:00',
]
```

```python
print('{:32} | {:32} | {:32}'.format('input', 'strict', 'flexible'))
for x in xs:
    try:
        one = pendulum.parse(x)
    except pendulum.parsing.ParserError:
        one = 'FAIL'
    try:
        two = pendulum.parse(x, strict=False)
    except pendulum.parsing.ParserError:
        two = 'FAIL'
    print('{:32} | {:32} | {:32}'.format(x, str(one), str(two)))
```

```python
ys = [pendulum.parse(x, strict=False) for x in xs]
for y in ys:
    print(y, '|', y.in_timezone('UTC'))
```
